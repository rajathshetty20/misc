```python
import qutip
```

```python
b = qutip.Bloch()

#add vectors without color
b.add_vectors([
    [1,1,1],
    [1,0,0],
    [1,1,2],
])

#add a vector with color
b.add_vectors([0,0,1],'grey')

#add multiple vectors with color
b.add_vectors([
    [2,2,1],
    [2,1,2]
],['k','cyan'])

b.show()
```

```python
import numpy as np

b.clear()

# add points without color
xp = [np.cos(th) for th in np.linspace(0, 2*np.pi, 20)]
yp = [np.sin(th) for th in np.linspace(0, 2*np.pi, 20)]
zp = np.zeros(20)
pnts = [xp, yp, zp]
b.add_points(pnts)

b.show()
```

```python
b.clear()

# add points with color in meth 's'
xp = [np.cos(th) for th in np.linspace(0, 2*np.pi, 20)]
yp = [np.sin(th) for th in np.linspace(0, 2*np.pi, 20)]
zp = np.zeros(20)
pnts = [xp, yp, zp]
b.add_points(pnts, meth="s", colors="k")

# add points with color in meth 'm'
zp = [1,-1,0]
xp = [0,0,0]
yp = [0,0,0]
pnts = [xp, yp, zp]
b.add_points(pnts, meth="m", colors=["cyan", "grey", "orange"])

b.show()
```
